```python
def crawl_email(memberEmails):
    driver.switch_to_frame("cafe_main")
    tbody = driver.find_element_by_css_selector(".article-board:not([id='upperArticleList'])")
    nicks = tbody.find_elements_by_css_selector(".p-nick > a")
    if not nicks:
        driver.quit()
        return False
    else:
        tmp = []
        for t in nicks:
            tmp.append(t.get_attribute('onclick'))
        [t.split(',')[1].replace("'",'').strip() for t in tmp]
        memberEmails += [f"{t.split(',')[1]}@naver.com".replace("'",'').strip() for t in tmp]
        time.sleep(5)
        return True
        
```

```python
try:
    pages = driver.find_elements_by_css_selector('div[class="prev-next"] > a[href]')
    pageUrl  = str(pages[0].get_attribute('href'))
    memberEmails=[]
    for c in range(1, 10001 ,1):    
        replaceUrl = pageUrl.replace('search.page=1','search.page=' + str(c))
        driver.get(replaceUrl)
        if crawl_email(memberEmails):
            continue
        else:
            break

except ValueError:
    print(e)
    
finally :
    driver.quit()
```
